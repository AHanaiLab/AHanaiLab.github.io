```python
m = folium.Map(location=location,tiles='cartodbpositron',zoom_start=10)

folium.Choropleth(
    geo_data=geojson,
    name='choropleth',
    data=data,
    columns=['@area', '昼夜間人口比率'],
    key_on='feature.properties.N03_007',
    fill_color='YlGnBu'
).add_to(m)

folium.LayerControl().add_to(m)

m.save(f'locations/{pref_romaji}_location.html')
```

```python
#適当な列番号（22)でマップしてみる
#columnsの引数には、新しく追加した列（col_name[47]を指定する！
# コピーして操作
df_tmp = data_prefecture.copy()


# 列番号22のデータを10分位化
df_tmp["区分"] = pd.qcut(
    df_tmp[col_name[22]],
    q=10,
    labels=False,  # 0〜9
    duplicates="drop"  # 値が少なくて重複する場合は自動調整
)


# Foliumマップ作成
m = folium.Map(location=location, tiles='cartodbpositron', zoom_start=10)
folium.Choropleth(
    geo_data=geojson,  # 対象県のGeoJSON
    name='choropleth',
    data=df_tmp,
    columns=[col_name[47], "区分"],  # col_name[47]は地域コード列
    key_on='feature.properties.N03_007',  # GeoJSONの自治体コードキー
    fill_color='YlGnBu'
).add_to(m)
folium.LayerControl().add_to(m)

m.save(f'junior_highschools/{pref_romaji}_'+col_name[22]+'.html')
```

```python
for i in range(1, len(col_name)-1):
    # データコピー
    df_tmp = data_prefecture.copy()

    # 10分位化
    df_tmp["区分"] = pd.qcut(
        df_tmp[col_name[i]],
        q=10,
        labels=False,  # 0〜9
        duplicates="drop"  # 値が少なくて重複する場合は自動調整
    )

    # Foliumマップ作成
    m = folium.Map(location=location, tiles='cartodbpositron', zoom_start=10)
    folium.Choropleth(
        geo_data=geojson,  # 対象県のGeoJSON
        name='choropleth',
        data=df_tmp,
        columns=[col_name[47], "区分"],  # col_name[47]は地域コード列
        key_on='feature.properties.N03_007',  # GeoJSONの自治体コードキー
        fill_color='YlGnBu'
    ).add_to(m)
    folium.LayerControl().add_to(m)

    # 保存
    output_path = f'{pref_romaji}_map/{pref_romaji}_{col_name[i]}_qcut.html'
    print(output_path)
    m.save(output_path)
```
